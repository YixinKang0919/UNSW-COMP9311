## 1. Relational Data Model Concepts
The relational data model is the most widely used data model for database
systems

Relational model has two styles of terminology:
* Mathematical: relation, tuple, attribute, ...
* data-oriented: table, record, field/column, ...

### Relational

A relation schema R is formally denoted by:

                                R(A1, A2, A3 ..., An)

where Ai denotes an attribute in R:

                            e.g., STUDENT(Name, Sid, Age, GPA)

#### Formal definitions
* A relation r of the relation schema R(A1, A2, ... , An) is denoted by r(R)
* r(R) is a set of n-tuples, i.e., r = {t1, t2, ... , tm}
* Each tuple t is an ordered list of values t = <v1, v2, v3 ..., vn> where each vi is an element of dom(Ai)

### Characteristic of Relations

#### Ordering of Tuples in a Relation
no order

#### Ordering of Values within a tuple
order is important!!!

#### Values and NULLs in the Tuples: each value in a tuple is an atomic value
* e.g.  "Benjamin Bayer" is a single string value( not two words)
* Person.Favourite_Fo

NULL
* Unknown
* Not available
* Does not apply

### Common Types of Integrity Constraints

#### Domain Constraints
####  Superkeys and keys
####  Null value constraint
####  Entity Integrity Constraint
* No primary key value can be NULL
* Applies to a single relation, but important in the context of multiple relations
####  Referential Integrity Constraint
* Between two relations, a tuple in one relation that refers to another relation must refer to an existing tuple in that relation
* describe references between relations(tables)
* are related to notion of a foreign key(FK), i.e.,a foreign key must refer from another relation -- primary key

### Updates and Dealing with Constraint Violation
>Follow the constraint!!!

#### Insertions
* You can't insert exsiting key
* You can't insert Null key
* You can't insert a foreign key where the primary key doesn't exist
#### Deletions
* You can't delete a primary key that is used in other relations as foreign key
#### Modifications
All the insertion and deletion rules

## 2. ER to Relational Data Model

### 2.1 Mapping Strong entity
Step 1: For each strong entity (not weak entity) type E, create a new relation R with
* Attributes : all simple attributes (and simple components of composite attributes) of E
* Key : key of E as the primary key for the relation. 

### 2.2 Mapping Weak entity
Step 2: For each weak entity type W with the owner entity type E, create a new relation R with
#### Attributes :
* all simple attributes (and simple components of composite attributes) of W,
* and include the primary key attributes of the relation derived from E as the foreign key.
#### Key of R:
* foreign key to E and partial key of W.

### 2.3 Mapping 1:1 Relationship Types
Step 3: For each 1:1 relationship type B, let E and F be the participating entity types. Let S and T be the corresponding relations.
* Choose one of S and T (let S be the one that participates totally if there is one).
* Add attributes from the primary key of T to S as a foreign key.
* Add all simple attributes (and simple components of composite attributes) of B as attributes of S.

### 2.4 Mapping 1:N Relationship Types
Step 4: For each 1:N relationship type B, let E and F be the participating entity types. Let S and T be the corresponding relations. E is the entity on the 1 side and F on the N side. 

### 2.5 Mapping M:N Relationship Types
Step 5: For each N:M relationship type B, let E and F be the participating entity types. Let S and T be the corresponding relations

Create a new relation R (cross-reference) with

#### Attributes: 
* Attributes from the key of S as foreign key,
* And attributes from the key of T as foreign key,
* And simple attributes, and simple components of composite attributes of relation B

#### Key:
* All attributes from the key of S and the key of T.

### 2.6 Mapping Multivalued Attributes
Step 6: For each multivalued attribute A, where A is an attribute of E, create a new relation R.

If A is a multivalued simple attribute,
* Attributes of R = Simple attribute A, and key of E as a foreign key

If A is a multivalued composite attribute,
* Attributes of R = All simple components of A, and key of E as a foreign key

### 2.7 Mapping N-ary Relationship Types
Step 7: For each n-ary relationship type (n > 2), create a new relation S with

#### Attributes:
* Include as foreign key attributes in S the primary keys of the relations that represent the participating entity types.
* Also include any simple attributes of the n-ary relationship type (or simple components of composite attributes) as attributes of S.
#### Key:
* All attributes from the primary keys of the participating entity types
